```python
import nest_asyncio
from pathlib import Path
from dotenv import load_dotenv
import os
import json

from IPython.display import Markdown
from llama_parse import LlamaParse
load_dotenv()
LLAMA_CLOUD = os.getenv('LLAMA_CLOUD')
```

# Второй вариант

```python
instruction = """The provided document is a scientific article. Try to parse it accurately"""
parser = LlamaParse(
    api_key=LLAMA_CLOUD,
    result_type="markdown",
    #parsing_instruction=instruction,
    max_timeout=5000,
)

llama_parse_documents = await parser.aload_data("data/pone.0083988.pdf")
     
```

```python
full_doc= ''.join([x.text for x in llama_parse_documents])
```

```python
Markdown(full_doc)
```

```python
document_path = Path("processed_data/pone.0083988.md")
with document_path.open("a") as f:
    f.write(full_doc)
```
